```python
df = pd.read_pickle('winback_data_500k_processed.pkl')
print(df.shape)
df.head()
```

```python
cols_to_drop = ['is_subscriber']
```

```python
df = df.drop(columns=cols_to_drop)
print(df.shape)
df.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, stratify=y,test_size=0.2, random_state=42)
```

```python
y_pred = model.predict_proba(x_test)

pr_auc = average_precision_score(y_test, y_pred[:,1])
print(pr_auc)

auc = roc_auc_score(y_test, y_pred[:,1])
print(auc)

y_pred_res = model.predict(x_test)

print(classification_report(y_test, y_pred_res))

metrics.f1_score(y_test, y_pred_res)



```

# Combo samping

```python
threshold = [0.5, 0.6, 0.7, 0.8, 0.9, 1]
u_threshold = [0.7, 0.8, 0.9, 1,1.5]
fs = 0
over_thresh = 0
under_thresh = 0
```

```python
for o in threshold:
    for u in u_threshold:
        if o == 1.5:
            continue
        over = RandomOverSampler(sampling_strategy=o)

        # first performing oversampling on minority class
        x_over, y_over = over.fit_resample(x_train, y_train)

        y_over = pd.Series(y_over)
        
        if u != 1.5 and u > o:
        # now combine undersampling 
            under = RandomUnderSampler(sampling_strategy=u) 
            x_combined_sampling, y_combined_sampling = under.fit_resample(x_over, y_over)
        else:
            x_combined_sampling, y_combined_sampling = x_over, y_over

        x_train_combo = pd.DataFrame(x_combined_sampling, columns = x_train.columns)
        y_train_combo = pd.Series(y_combined_sampling)

        # name the series
        y_train_combo.name = "winback_target"
        
        model = lgbm.LGBMClassifier(random_state=42)
        model.fit(x_train_combo, y_train_combo)
        
        y_pred = model.predict_proba(x_test)
        y_pred_res = model.predict(x_test)
        
        f = metrics.f1_score(y_test, y_pred_res)
        
        if f > fs:
            fs = f
            over_thresh = o
            under_thresh = u

print(fs)
print(over_thresh)
print(under_thresh)
```

```python
over = RandomOverSampler(sampling_strategy=0.7)
under = RandomUnderSampler(sampling_strategy=1) 

# first performing oversampling on minority class
x_over, y_over = over.fit_resample(x_train, y_train)

y_over = pd.Series(y_over)
print("Oversampled \n", y_over.value_counts())

# now combine undersampling 
x_combined_sampling, y_combined_sampling = under.fit_resample(x_over, y_over)

x_train_combo = pd.DataFrame(x_combined_sampling, columns = x_train.columns)
y_train_combo = pd.Series(y_combined_sampling)

# name the series
y_train_combo.name = "winback_target"
```

```python
start = time.time()

model = lgbm.LGBMClassifier(random_state=42)
model.fit(x_train_combo, y_train_combo)

end = time.time()
print(end - start)
```

```python
def lift_precision_recall_fscore(percentile, prob_list,  y_pred, y):
    
    
    rows = y_pred.shape[0]
    res1 = pd.DataFrame({'prob':prob_list, 'precision':prob_list, 'recall':prob_list, 'fscore':prob_list})
    
    for pr in prob_list : 
        prf = metrics.precision_recall_fscore_support(y, [int(i > pr) for i in  y_pred], average='binary')
        res1.loc[res1.prob == pr,'precision'] = prf[0]
        res1.loc[res1.prob == pr,'recall'] = prf[1]
        res1.loc[res1.prob == pr,'fscore'] = prf[2]    
    
    print(res1)
```

```python
y_pred_train = model.predict_proba(x_train)
y_pred_test = model.predict_proba(x_test)

y_pred_train = y_pred_train[:,1]
y_pred_test = y_pred_test[:,1]


lift_list = [0.05, 0.1, 0.2, 0.5, 0.7]
metrics_prob = [0.05, 0.1,0.15,0.2,0.25, 0.3,0.35,0.4,0.45, 0.5, 0.55, 0.7,0.9]
target_col = "last_touch_sub"

print('Train set thresholds')
lift_precision_recall_fscore(lift_list, metrics_prob, y_pred_train, y_train)
print()
print('Test set thresholds')
lift_precision_recall_fscore(lift_list, metrics_prob, y_pred_test, y_test)
print()
```
